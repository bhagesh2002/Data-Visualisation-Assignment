3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and

Heat map for sales data:

```python
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

fig = go.Figure(data=go.Heatmap(
    z=pivot_table.values,
    x=pivot_table.columns,
    y=pivot_table.index,
    colorscale='Viridis'
))

fig.update_layout(
    title='Heatmap of Sales Data',
    xaxis_title='Month',
    yaxis_title='Day'
)

fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot in the notebook
output_notebook()
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
import numpy as np
from bokeh.palettes import Viridis256

np.random.seed(42)
x = np.random.rand(100) * 10  
y = np.random.rand(100) * 10  
sizes = np.random.rand(100) * 30 + 10  
colors = np.random.rand(100)  

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Scatter Plot with Random Sizes and Colors", x_axis_label='x', y_axis_label='y')

color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

p.scatter('x', 'y', size='sizes', color=color_mapper, source=source, fill_alpha=0.6)

# Show the plot in the notebook
output_notebook()
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral4

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')

p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral4, factors=fruits))

p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot in the notebook
output_notebook()
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure, show, output_notebook

data_hist = np.random.randn(1000)  # Generate 1000 random numbers from a normal distribution
hist, edges = np.histogram(data_hist, bins=30)  # Compute the histogram

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

output_notebook()
show(p)
```

```python
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.palettes import Viridis256

# Prepare the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

color_mapper = LinearColorMapper(palette=Viridis256, low=np.min(data_heatmap), high=np.max(data_heatmap))

p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y',
           x_range=(0, 1), y_range=(0, 1), tools="hover", tooltips=[("Value", "@image")])

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0), title="Scale")
p.add_layout(color_bar, 'right')

output_notebook()
show(p)
```
